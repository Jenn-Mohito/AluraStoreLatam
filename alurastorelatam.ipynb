```python
tiendas = {
    "Tienda 1": tienda_1,
    "Tienda 2": tienda_2,
    "Tienda 3": tienda_3,
    "Tienda 4": tienda_4
}
def resumen_eda(df, nombre_tienda):
    return {
        "Tienda": nombre_tienda,
        "Total Registros": df.shape[0],
        "Filas Duplicadas": df.duplicated().sum(),
        "Valores Nulos": df.isnull().sum().sum(),
        "Precio Promedio": round(df["Precio"].mean(), 2),
        "Precio Mínimo": df["Precio"].min(),
        "Precio Máximo": df["Precio"].max(),
        "Costo Envío Promedio": round(df["Costo de envío"].mean(), 2),
        "Costo Envío Mínimo": df["Costo de envío"].min(),
        "Costo Envío Máximo": df["Costo de envío"].max(),
        "Calificación Promedio": round(df["Calificación"].mean(), 2),
        "Calificación Mínima": df["Calificación"].min(),
        "Calificación Máxima": df["Calificación"].max(),
        "Productos Únicos": df["Producto"].nunique(),
        "Categorías Únicas": df["Categoría del Producto"].nunique(),
        "Vendedores Únicos": df["Vendedor"].nunique(),
        "Fecha Inicio": df["Fecha de Compra"].min(),
        "Fecha Fin": df["Fecha de Compra"].max(),
        "Tipo de Fecha": df["Fecha de Compra"].dtypes
    }
# Crear lista con resumen de cada tienda
all_eda_data = []
for nombre_tienda, df in tiendas.items():
    all_eda_data.append(resumen_eda(df, nombre_tienda))
# Crear DataFrame resumen
summary_df = pd.DataFrame(all_eda_data)

summary_df
```

# 2. Ventas por categoría

```python
tiendas = {
    "Tienda 1": tienda_1,
    "Tienda 2": tienda_2,
    "Tienda 3": tienda_3,
    "Tienda 4": tienda_4
}
# Usamos groupby + size y luego unimos todo con concat
ventas_por_categoria_df = pd.concat(
    {
        tienda: df.groupby("Categoría del Producto").size()
        for tienda, df in tiendas.items()
    },
    axis=1
).fillna(0).astype(int)
# Creamos la columna Total por Categoría
ventas_por_categoria_df["Total por Categoría"] = ventas_por_categoria_df.sum(axis=1)
# Ordenamos por total para el análisis
ventas_por_categoria_df = ventas_por_categoria_df.sort_values(
    by="Total por Categoría",
    ascending=False
)
# reseteamos indice y renombramos columna Categoría del Producto
ventas_por_categoria_df = ventas_por_categoria_df.reset_index()
ventas_por_categoria_df.rename(
    columns={"Categoría del Producto": "Categoría de Productos"},
    inplace=True
)
```

```python
display(ventas_por_categoria_df)
```

```python
import os

# Graficamos con un Heatmap
heatmap_data = ventas_por_categoria_df.set_index(
    "Categoría de Productos"
)[["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]]
plt.figure(figsize=(12, 8))

sns.heatmap(
    heatmap_data,
    annot=True,
    fmt="d",
    cmap="rocket",
    linewidths=2,
    linecolor='#f0f0f0',
    cbar_kws={"label": "Volumen de Ventas"}
)

plt.title("Heatmap de Ventas por Categoría y Tienda", fontsize=14)
plt.xlabel("Tiendas")
plt.ylabel("Categoría de Productos")
plt.tight_layout()

# Crear la carpeta 'imgs' si no existe
if not os.path.exists('imgs'):
    os.makedirs('imgs')

plt.savefig("imgs/heatmap_ventas_por_categoria_y_tienda.png")
plt.show()
```

##### Calificación promedio de los últimos 12 meses

```python
#Dataframe consolidado
tiendas_dfs = {
    'tienda_1': tienda_1,
    'tienda_2': tienda_2,
    'tienda_3': tienda_3,
    'tienda_4': tienda_4
}

# Add 'Tienda' column to each DataFrame and collect them in a list
lista_dfs_para_concatenar = []
for nombre_tienda, df in tiendas_dfs.items():
    df_copy = df.copy() # Create a copy to avoid modifying original DataFrames directly if they are used later
    df_copy['Tienda'] = nombre_tienda
    lista_dfs_para_concatenar.append(df_copy)

# Concatenate all DataFrames
df_consolidado = pd.concat(lista_dfs_para_concatenar, ignore_index=True)

# Create 'Mes-Año' column
df_consolidado['Mes-Año'] = df_consolidado['Fecha de Compra'].dt.strftime('%Y-%m')

print("DataFrame consolidado (df_consolidado) creado con éxito.")
print(f"Dimensiones de df_consolidado: {df_consolidado.shape}")
print("Primeras 5 filas de df_consolidado:")
display(df_consolidado.head())
```

```python
#Gráfico promedio de los últimos 12 meses
ventas_mensuales = df_consolidado.groupby(['Mes-Año', 'Tienda'])['Precio'].sum().reset_index()

# 2. Configuración de estilo y tamaño
sns.set_theme(style="whitegrid")
plt.figure(figsize=(15, 7))

# 3. Creación del gráfico de barras agrupadas
ax = sns.barplot(
    data=ventas_mensuales,
    x='Mes-Año',
    y='Precio',
    hue='Tienda',
    palette='viridis'
)

# 4. Personalización de títulos y etiquetas
plt.title('Comparativo Mensual de Ventas por Sede', fontsize=16, fontweight='bold', pad=25)
plt.xlabel('Periodo (Mes-Año)', fontsize=12)
plt.ylabel('Facturación Total (USD)', fontsize=12)

# 5. Ajustes de formato
plt.xticks(rotation=45)
plt.ticklabel_format(style='plain', axis='y')
plt.legend(title='Tiendas', bbox_to_anchor=(1.05, 1), loc='upper left')

# 6. Limpieza y guardado
sns.despine()
plt.tight_layout()
plt.savefig('imgs/ventas_mensuales_barras.png')
plt.show()
```

Los niveles de satisfacción en Alura Store reflejan una notable solidez y consistencia, con calificaciones promedio que se mantienen en un rango alto y estrecho entre 3.98 y 4.05 puntos. Aunque la Tienda 3 lidera ligeramente el índice de satisfacción con la puntuación más alta (4.05), la mínima diferencia de apenas 0.07 puntos entre todas las sucursales confirma que la excelencia en el servicio es un estándar compartido. # 4. Productos más y menos vendidos

```python
top_products_data = [] #inicializamos lista para almacenar los top 5 productos
bottom_products_data = [] # inicializamos lista para almacenar los 5 productos menos vendidos

for i, tienda_df in enumerate([tienda_1, tienda_2, tienda_3, tienda_4]):
    store_name = f'tienda_{i+1}'
    # Contar la frecuencia de cada producto
    product_counts = tienda_df['Producto'].value_counts().reset_index()
    product_counts.columns = ['Producto', 'Cantidad de Ventas']

    # Seleccionar los 5 productos más vendidos
    top_5_products = product_counts.head(5).copy()
    top_5_products.loc[:, 'Tienda'] = store_name
    top_5_products.loc[:, 'Tipo'] = 'Más Vendidos'
    top_products_data.append(top_5_products)

    # Seleccionar los 5 productos menos vendidos
    bottom_5_products = product_counts.tail(5).copy()
    bottom_5_products.loc[:, 'Tienda'] = store_name
    bottom_5_products.loc[:, 'Tipo'] = 'Menos Vendidos'
    bottom_products_data.append(bottom_5_products)

# Concatenar todos los DataFrames de productos principales (needed for visualization cell)
all_top_products_df = pd.concat(top_products_data, ignore_index=True)
all_bottom_products_df = pd.concat(bottom_products_data, ignore_index=True)

# Mostrar los top 5 productos más vendidos por tienda
print("--- Top 5 Productos Más Vendidos por Tienda ---")
for df in top_products_data:
    store_name = df['Tienda'].iloc[0]
    print(f"\n--- {store_name} (Más Vendidos) ---")
    display(df[['Producto', 'Cantidad de Ventas']])
# Mostrar los 5 productos menos vendidos por tienda
print("\n--- Top 5 Productos Menos Vendidos por Tienda ---")
for df in bottom_products_data:
    store_name = df['Tienda'].iloc[0]
    print(f"\n--- {store_name} (Menos Vendidos) ---")
    display(df[['Producto', 'Cantidad de Ventas']])
```

```python
# Consolidamos todos los datos en dos DataFrames globales
df_top_global = pd.concat(top_products_data).groupby('Producto')['Cantidad de Ventas'].sum().sort_values(ascending=False).head(10).reset_index()
df_bottom_global = pd.concat(bottom_products_data).groupby('Producto')['Cantidad de Ventas'].sum().sort_values(ascending=True).head(10).reset_index()

# GRÁFICO 1: TOP 10 GLOBAL (LOS MÁS VENDIDOS)
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")
ax1 = sns.barplot(data=df_top_global, x='Cantidad de Ventas', y='Producto', palette='viridis')

plt.title('Líderes de Ventas: Top 10 Productos Globales', fontsize=15, fontweight='bold', pad=20)
plt.xlabel('Total Unidades Vendidas')
plt.ylabel('')
ax1.bar_label(ax1.containers[0], padding=5, fontweight='bold')
sns.despine()
plt.tight_layout()
plt.savefig('imgs/top_global_ventas.png')
plt.show()

#GRÁFICO 2: BOTTOM 10 GLOBAL (LOS MENOS VENDIDOS)
plt.figure(figsize=(12, 6))
ax2 = sns.barplot(data=df_bottom_global, x='Cantidad de Ventas', y='Producto', palette='flare')

plt.title('Oportunidades de Mejora: 10 Productos con Menor Rotación', fontsize=15, fontweight='bold', pad=20)
plt.xlabel('Total Unidades Vendidas')
plt.ylabel('')
ax2.bar_label(ax2.containers[0], padding=5, fontweight='bold')
sns.despine()
plt.tight_layout()
plt.savefig('imgs/bottom_global_ventas.png')
plt.show()
```
